```python
# import
import datetime
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
# warnings
import warnings
# With np.nanmean, for all-NaN slices, NaN is returned and a RuntimeWarning is raised.
warnings.filterwarnings("ignore","Mean of empty slice", RuntimeWarning)
# With np.nanstd, for all-NaN slices or slices with zero degrees of freedom, NaN is returned
#and a RuntimeWarning is raised.
warnings.filterwarnings("ignore","Degrees of freedom <= 0 for slice.", RuntimeWarning)
```

```python
# Parameters

#track = '222' #pass number

fic_20hz = 'ESACCI-SEALEVEL-L3-SLA-MED_SEA-MERGED-20190927-JA-222-fv01.nc' # X-TRACK/ALES 20 Hz

# period / J2 / calendar is Julian
tref = datetime.datetime(1950,1,1)
tstart = datetime.datetime(2010,1,1)
tend = datetime.datetime(2015,12,31)
```

```python
# read SLA files

xt20hz = xr.open_dataset(fic_20hz, decode_times=False)
sla20hz = xt20hz['sla'].data
lat20hz = xt20hz['lat'].data
lon20hz = xt20hz['lon'].data
time20hz = np.nanmean(xt20hz['time'].data, axis=0)
mdt20hz = xt20hz['mdt_cnes_cls18'].data
d2c20hz = xt20hz['dist_to_coast_gshhs'].data

dx20,dt20 = sla20hz.shape
```

```python
# fill in the holes (nan value) in the time vector
# transformation of time (julian day) into date


for t in range(len(time20hz)): 
    if np.isnan(time20hz[t]): 
        time20hz[t] = time20hz[t-1]+9.9156 
td20hz=np.array([tref + datetime.timedelta(days=t) for t in time20hz])


id20start = min(np.where(td20hz > tstart)[0])
id20end = max(np.where(td20hz < tend)[0])

td20hz = td20hz[id20start:id20end]
sla20hz = sla20hz[:,id20start:id20end]
```

```python
dist20hz=d2c20hz
```
